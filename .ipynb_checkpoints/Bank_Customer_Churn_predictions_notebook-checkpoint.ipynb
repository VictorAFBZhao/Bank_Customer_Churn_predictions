#### Hypothesis Testing:

Hypothesis Testing for the difference in mean age between customers who churned (churn=1) and those who stayed (churn=0):
This hypothesis test can help determine if age has a significant impact on customer churn.

```python
from scipy import stats
```

```python
churned_age =  churn[churn['churn'] == 1]['age']
not_churned_age =  churn[churn['churn'] == 0]['age']
```

```python
t_stat, p_value = stats.ttest_ind(churned_age, not_churned_age)
print(f'T-stat:{t_stat}, P_value {p_value}')

if p_value < 0.05:
    print("The difference in age between customers who churned and those who is statistically significant.")
else:
    print("There is no statistical evidence of a difference in age between customers who churned and those who stayed.")
```
